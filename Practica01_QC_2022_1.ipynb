$$\newcommand{\ket}[1]{\left|{#1}\right\rangle}$$
$$\newcommand{\bra}[1]{\left\langle{#1}\right|}$$
$$\newcommand{\braket}[2]{\left\langle{#1}\middle|{#2}\right\rangle}$$

> **Operaciones con vectores:** De las operaciones descritas en el artículo se ejemplificarán la suma de vectores y la multiplicación por un escalar, ya que son los casos de más uso en la computación cuántica. Partimos de la definición de 2 vectores en $R^{n}$, la definición de la suma y la definición de multiplicación por un escalar $k$:

$$\ket{a}=\begin{bmatrix} a_{1}\\ a_{2} \\ \vdots \\ a_{n}\end{bmatrix}, \ket{b}=\begin{bmatrix} b_{1}\\ b_{2} \\ \vdots \\ b_{n} \end{bmatrix},\ket{a}+\ket{b}=\begin{bmatrix} a_{1}+b_{1}\\ a_{2}+b_{2} \\ \vdots \\ a_{n}+b_{n} \end{bmatrix}$$

$$k\ket{a}= \begin{bmatrix} ka_{1}\\ ka_{2} \\ \vdots \\ ka_{n}\end{bmatrix}$$

> **Ejemplo 1**:
$$\ket{a}=\begin{bmatrix} 2\\ 3\end{bmatrix}, \ket{b}=\begin{bmatrix} -1 \\ 6\end{bmatrix}, \ket{a}+\ket{b}=\begin{bmatrix} 1 \\ 9 \end{bmatrix}$$

> **Ejemplo 2**:
$$\ket{a}=\begin{bmatrix} 8\\ -2\end{bmatrix}, {k}=5, k\ket{a}=\begin{bmatrix} 40 \\ -10 \end{bmatrix}$$

> **Bases ortonormales**: Una base es un grupo de vectores que pueden generar a todos los vectores en un determinado espacio vectorial, es decir, para el caso de 2 dimensiones, una base será un grupo de vectores de dos dimensiones que mediante suma y multiplicación por escalar pueden dar como resultado cualquier vector. Una base debe ser ortogonal por definición, es decir, los vectores que componen la base tienen que ser perpendiculares entre sí. Además si la magnitud de cada uno de los vectores de una base es 1, a esta base se le llama normal, así que al unir ambos nombres se obtiene el apelativo de ortonormal. Para demostrar que son ortogonales se realiza el producto interno entre ambos, y para demostrar que son normales se realiza el producto interno de cada uno consigo mismo.

> **Ejemplo 3**:
$$\ket{a}=\begin{bmatrix} 4/\sqrt{17}\\ -1/\sqrt{17}\end{bmatrix}, \ket{b}=\begin{bmatrix} 2/\sqrt{68} \\ 8/\sqrt{68}\end{bmatrix}$$

$$\braket{a}{b}=\begin{bmatrix} 4/\sqrt{17}, -1/\sqrt{17}\end{bmatrix} \cdot \begin{bmatrix} 2/\sqrt{68} \\ 8/\sqrt{68}\end{bmatrix}=0$$

$$\braket{a}{a}=\begin{bmatrix} 4/\sqrt{17} , -1/\sqrt{17}\end{bmatrix} \cdot \begin{bmatrix} 4/\sqrt{17} \\ -1/\sqrt{17}\end{bmatrix}=1$$

$$\braket{b}{b}=\begin{bmatrix} 2/\sqrt{68} , 8/\sqrt{68}\end{bmatrix} \cdot \begin{bmatrix} 2/\sqrt{68} \\ 8/\sqrt{68}\end{bmatrix}=1$$

> De esta manera se concluye que la pareja de vectores $\ket{a}$ y $\ket{b}$ son una base ortonormal para $R^{2}$.

```python
import numpy as np
# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile, Aer, assemble, IBMQ
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from qiskit.visualization import plot_histogram, plot_bloch_vector
from math import sqrt, pi
from ibm_quantum_widgets import *
from qiskit.providers.aer import QasmSimulator
from qiskit.visualization import array_to_latex, plot_bloch_multivector
from qiskit.quantum_info import Statevector

# Loading your IBM Quantum account(s)
# provider = IBMQ.load_account()

qc = QuantumCircuit(1)  # Create a quantum circuit with one qubit
initial_state = [0,1]   # Define initial_state as |1>
qc.initialize(initial_state, 0) # Apply initialisation operation to the 0th qubit
qc.draw(output='mpl', style={'backgroundcolor': '#EEEEEE'})  # Let's view our circuit
```

```python
sim = Aer.get_backend('aer_simulator')  # Tell Qiskit how to simulate our circuit

qc = QuantumCircuit(1)  # Create a quantum circuit with one qubit
initial_state = [0,1]   # Define initial_state as |1>
qc.initialize(initial_state, 0) # Apply initialisation operation to the 0th qubit
qc.save_statevector()   # Tell simulator to save statevector
qobj = assemble(qc)     # Create a Qobj from the circuit for the simulator to run
result = sim.run(qobj).result() # Do the simulation and return the result

out_state = result.get_statevector()
print(out_state) # Display the output state vector

qc.measure_all()
qc.draw(output='mpl', style={'backgroundcolor': '#EEEEEE'})

```

```python
qobj = assemble(qc)
result = sim.run(qobj).result()
counts = result.get_counts()
plot_histogram(counts)
```

```python
initial_state = [1/sqrt(2), 1j/sqrt(2)]  # Define state |q_0>

qc = QuantumCircuit(1) # Must redefine qc
qc.initialize(initial_state, 0) # Initialize the 0th qubit in the state `initial_state`
qc.save_statevector() # Save statevector
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector() # Execute the circuit
print(state)           # Print the result

qobj = assemble(qc)
results = sim.run(qobj).result().get_counts()
plot_histogram(results)
```

## 5.3 **Conclusiones**<br>
* Los sistemas cuánticos, al no ser determinísticos, y al obedecer el principio de incertidumbre de Heisenberg, permiten en principio realizar mediciones de un estado cuántico solo a lo largo de uno de los ejes del sistema de coordenadas usado; pero gracias a las compuertas que permiten rotar estos estados, es posible hacer mediciones desde otros ejes, lo cual permite obtener una información más amplia y compleja respecto al qubit.<br>
